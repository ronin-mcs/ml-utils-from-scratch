```python
# Посчитаем количество выживших и их пропорцию
survived = df[df["Survived"]==1].count()
print("Survived count: ", survived["Survived"])
not_survived = df[df["Survived"]==0].count()
print("Not survived count: ", not_survived["Survived"])

survived_perc = survived["Survived"] / (survived["Survived"] + not_survived["Survived"]) * 100

print("Survived percentage: ", survived_perc)
```

```python
# Взгляем на пропорцчию выживших в зависимости от класса пассажира
# df.groupby('Sex')['Survived'].mean()

# print(df["Sex"].unique()[0])
survied_female_perc = df[(df["Sex"] == df["Sex"].unique()[1]) & (df["Survived"] == 1)].shape[0] / df[df["Sex"] == df["Sex"].unique()[1]].shape[0]
print("percentage of survived female: ", survied_female_perc * 100, "%")

survied_male_perc = df.groupby('Sex')['Survived'].mean().iloc[1]
print("percentage of survived male: ", survied_male_perc * 100, "%")

print(df.groupby('Pclass')['Survived'].mean())
```

```python
# Чистим вилкой
df_cleaned = df.drop(columns=["PassengerId", "Name", "Ticket", "Cabin"])
df_cleaned = df_cleaned.fillna({
    "Age": df_cleaned["Age"].median(),
    "Embarked": df_cleaned["Embarked"].mode()[0]
})
df_cleaned = df_cleaned.dropna(subset=["Fare"])
```

```python
df_cleaned = pd.get_dummies(df_cleaned, columns=['Sex', 'Embarked'], drop_first=True)
```
